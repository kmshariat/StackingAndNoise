# Increasing SNR of Periodic Signals via Stacking Method

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

## Dataset 1

```python
#defining a gaussian curve to be our actual signal
def gaussian(t, sig, mu):
    return (1/(sig*np.sqrt(2*np.pi))) * np.exp(-0.5*((t-mu)/sig)**2)

#defining the time axis with three piece
#time1 and time 3 representing the time axis for noise
#time2 representing the time axis for actual signal

time1 = np.arange(0,40,1)
time2 = np.arange(40,60,1)
time3 = np.arange(60,100,1)

#randomly generating noise and calculating the values of the actual signal
value1_1 = np.array([random.uniform(0,2) - 1 for _ in range(40)])
value2_1 = np.array(100*gaussian(time2, 5, 50)-1)
value3_1 = np.array([random.uniform(0,2)-1 for _ in range(40)])

#finding best fit line with degree 3 curve for noise
x_1 = np.polyfit(time1, value1_1,3)
y_1 = np.polyfit(time3, value3_1,3)


plt.title("Pseudo Signal 1")
plt.xlabel("Time ($s$)")
plt.ylabel("Intensity")

#plotting
plt.scatter(time1,value1_1, marker = '+', color = 'gray', label = 'Random Noise')
plt.plot(time1, x_1[0]*time1**3 + x_1[1]*time1**2 + x_1[2]*time1 + x_1[3] )

plt.plot(time2, value2_1, marker = '.', label = 'Signal Data')

plt.scatter(time3, value3_1, marker = '+', color = 'gray')
plt.plot(time3, y_1[0]*time3**3 + y_1[1]*time3**2 + y_1[2]*time3 + y_1[3])

plt.legend()
plt.show()
```

## Dataset 2

```python
#randomly generating noise and calculating the values of the actual signal
value1_2 = np.array([random.uniform(0, 2)-1 for _ in range(40)])
value2_2 = np.array(100*gaussian(time2, 5, 50)-1)
value3_2 = np.array([random.uniform(0,2)-1 for _ in range(40)])

#finding best fit line with degree 3 curve for noise
x_2 = np.polyfit(time1, value1_2,3)
y_2 = np.polyfit(time3, value3_2,3)

plt.title("Pseudo Signal 2")
plt.xlabel("Time ($s$)")
plt.ylabel("Intensity")

#plotting
plt.scatter(time1,value1_2, marker = '+', color = 'gray', label = 'Random Noise')
plt.plot(time1, x_2[0]*time1**3 + x_2[1]*time1**2 + x_2[2]*time1 + x_2[3] )

plt.plot(time2, value2_2, marker = '.', label = 'Signal Data')

plt.scatter(time3, value3_2, marker = '+', color = 'gray')
plt.plot(time3, y_2[0]*time3**3 + y_2[1]*time3**2 + y_2[2]*time3 + y_2[3])

plt.legend()
plt.show()
```

## Dataset 3

```python
#randomly generating noise and calculating the values of the actual signal
value1_3 = np.array([random.uniform(0, 2)-1 for _ in range(40)])
value2_3 = np.array(100*gaussian(time2, 5, 50)-1)
value3_3 = np.array([random.uniform(0,2)-1 for _ in range(40)])

#randomly generating noise and calculating the values of the actual signal
x_3 = np.polyfit(time1, value1_3,3)
y_3 = np.polyfit(time3, value3_3,3)

plt.title("Pseudo Signal 3")
plt.xlabel("Time ($s$)")
plt.ylabel("Intensity")

#plotting
plt.scatter(time1,value1_3, marker = '+', color = 'gray', label = 'Random Noise')
plt.plot(time1, x_3[0]*time1**3 + x_3[1]*time1**2 + x_3[2]*time1 + x_3[3])

plt.plot(time2, value2_3, marker = '.', label = 'Signal Data')

plt.scatter(time3, value3_3, marker = '+', color = 'gray')
plt.plot(time3, y_3[0]*time3**3 + y_3[1]*time3**2 + y_3[2]*time3 + y_3[3])

plt.legend()
plt.show()
```

## Stacked Output for 3 Dataset

```python
#best fit line for the stacked noise signals
x_f = np.polyfit(time1, (value1_1+value1_2+value1_3)/3, 3)
y_f = np.polyfit(time3, (value3_1+value3_2+value3_3)/3, 3)

#stacking and plotting
plt.scatter(time1, (value1_1+value1_2+value1_3)/3, marker="+", label="Random Noise")
plt.plot(time1, x_f[0]*time1**3 + x_f[1]*time1**2 + x_f[2]*time1 + x_f[3])

plt.plot(time2, (value2_1+value2_2+value2_3)/3, marker=".", label="Actual Signal")

plt.scatter(time3, (value3_1+value3_2+value3_3)/3, marker="+")
plt.plot(time3, y_f[0]*time3**3 + y_f[1]*time3**2 + y_f[2]*time3 + y_f[3])

plt.title('Stacked Signal')
plt.xlabel('time ($s$)')
plt.ylabel('Intensity')

plt.legend()
plt.show()
```
